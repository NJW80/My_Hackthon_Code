```python
pre_xmas = Mid_fwd[Mid_fwd['kickoff_time'] < '2018-12-25']
post_xmas = Mid_fwd[Mid_fwd['kickoff_time'] > '2018-12-25']

pre_xmas_players = np.array(pre_xmas['player'].unique())
post_xmas_players = np.array(post_xmas['player'].unique())

print(len(pre_xmas_players))
print(len(post_xmas_players))

intersect_players = np.intersect1d(pre_xmas_players, post_xmas_players, assume_unique = True)
print(len(intersect_players))

pre_xmas = pre_xmas[pre_xmas['player'].isin(intersect_players)]
post_xmas = post_xmas[post_xmas['player'].isin(intersect_players)]

pre_xmas_players = np.array(pre_xmas['player'].unique())
post_xmas_players = np.array(post_xmas['player'].unique())

print(len(pre_xmas_players))
print(len(post_xmas_players))
```

```python
def group_pos_player(df):
    df_by_pos_player = df.groupby(['position','player']).agg({'bonus': 'mean',
                                                              'bps': 'mean',
                                                              'total_points': 'mean'})
    return df_by_pos_player

pre_xmas_by_player = group_pos_player(pre_xmas)
post_xmas_by_player = group_pos_player(post_xmas)

print(pre_xmas_by_player.head(3))
print()
print(post_xmas_by_player.head(3))
```

```python
pre_post_df = pre_xmas_by_player.join(
    post_xmas_by_player,
    lsuffix='_pre',
    rsuffix='_post'
)
pre_post_df.head(10)
```

```python
pre_post_df['improvement_index'] = pre_post_df.apply(lambda row: row['total_points_post'] / row['total_points_pre'], axis=1)
pre_post_df['improvement_index'] = pre_post_df['improvement_index'].fillna(0)
pre_post_df
```
